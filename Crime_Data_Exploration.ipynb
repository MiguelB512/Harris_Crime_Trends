```python
# value counts of reporting agencies by year
df['Year'].value_counts()
```

```python
# value count of agencies per county per year
agencies = df.groupby('County')['Year'].value_counts()
agencies
```

```python
# Create agencies df
# shows the number of agencies reporting index crime rates each year
agencies_df = pd.DataFrame(agencies)
agencies_df.head(15)
```
