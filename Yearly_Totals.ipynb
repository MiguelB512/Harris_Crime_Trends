```python
# Import dependencies
import pandas as pd
```

```python
# Import Data
yearly_df = pd.read_csv("Resources/CSV/Texas-Crime-Index-2015-2020.csv")
yearly_df.head()
```

```python
# Rename total column
yearly_df.rename(columns={'Total':'Total_Crimes'}, inplace=True)
```

```python
# Make new DF's based on violent and non-violent offenses
nonviolent_offenses = yearly_df[["Auto Theft","Larceny","Burglary"]]
violent_offenses = yearly_df[["Rape", "Murder","Assault"]]
```

```python
# Sum of nonviolent_offenses
nonviolent_offenses = nonviolent_offenses.sum(axis=1)
nonviolent_offenses
```

```python
# Sum of violent_offenses
violent_offenses = violent_offenses.sum(axis=1)
violent_offenses
```

```python
# Add the series back to the main Dataframe
yearly_df['Violent_Offenses'] = violent_offenses
yearly_df['NonViolent_Offenses'] = nonviolent_offenses
```

```python
# Group crimes by year, replace spaces with underscores and reorder columns
yearly_df.columns = yearly_df.columns.str.replace(' ','_')
yearly_df.drop('Population', axis = 1, inplace = True)
yearly_df = yearly_df.groupby('Year').sum()
yearly_df = yearly_df[['Murder', 'Rape', 'Robbery', 'Assault', 'Burglary', 'Larceny',
       'Auto_Theft', 'Violent_Offenses', 'NonViolent_Offenses', 'Total_Crimes']]
yearly_df
```

```python
# Export County Summary to CSV
yearly_df.to_csv('Resources/CSV/yearly_totals.csv', index=True)
```
