## Check CSV Files

```python
# Load Oct-Dec 2021 DF
oct_dec_2021_df = pd.read_csv("Oct_Dec_2021_Houston_Police_Department_report.csv")
oct_dec_2021_df.head()
```

```python
# Load Jan-Mar 2022 DF
jan_mar_2022_df = pd.read_csv("Jan_Mar_2022_Houston_Police_Department_report.csv")
jan_mar_2022_df.head()
```

```python
# Load Apr-Jun 2022 DF
apr_jun_2022_df = pd.read_csv("Apr_Jun_2022_Houston_Police_Department_report.csv")
apr_jun_2022_df.head()
```

```python
# Load Jul-Sep 2022 DF
jul_sep_2022_df = pd.read_csv("Jul_Sep_2022_Houston_Police_Department_report.csv")
jul_sep_2022_df
```

```python
# Merging 0ct and Jan Dfs
merged_df = oct_dec_2021_df.append(jan_mar_2022_df, ignore_index = True)
merged_df
```

```python
# Merge merged_df and July
merged_df = merged_df.append(jul_sep_2022_df, ignore_index = True)
merged_df
```
