```python
# Importing crime data CSV
crime_df = pd.read_csv('Resources/CSV/crime_data.csv')
harris_df = pd.read_csv('Resources/CSV/harris_crime_data.csv')
outer_hou_df = pd.read_csv('Resources/CSV/outter_hou_counties.csv')
```

```python
# Apply default sns theme
sns.set_theme(style="darkgrid")

# Crate a visualization
sns.relplot(
    data = outer_hou_df,
    x = "Year", 
    y = "Population",
    hue = "County", 
    markers = True,
    style = 'County',
    size = 'Population'
)

plt.savefig('Plots/County_Plot.jpg', dpi=100)
plt.show(sns)
```
