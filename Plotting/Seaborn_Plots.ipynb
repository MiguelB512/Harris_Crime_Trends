# Seaborn Visualization

### Import Dependencies

```python
# Import dependencies
import os 
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

### Load CSV Files

```python
# Importing crime data CSV
crime_df = pd.read_csv('../Resources/CSV/crime_data.csv')
harris_df = pd.read_csv('../Resources/CSV/harris_crime_data.csv')
outer_hou_df = pd.read_csv('../Resources/CSV/outer_hou_counties.csv')
```

## Plots

```python
# Apply default sns theme
sns.set_theme(style="darkgrid")

# Crate a visualization
sns.relplot(
    data = crime_df,
    y = "Total_Crime", 
    x = "Population",
    hue = "Year"
)
plt.show(sns)
```

```python
sns.relplot(
    data = crime_df, 
    kind="line",
    x="Population", 
    y="Total_Crime", 
    hue="Year",
    col="County", 
    col_wrap=5,
    estimator=None,
)
plt.show(sns)
```
