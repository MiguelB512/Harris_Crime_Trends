# Seaborn Visualization

### Import Dependencies

```python
# Import dependencies
import os 
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
# %matplotlib inline
```

### Load CSV Files

```python
# Importing crime data CSV
crime_df = pd.read_csv('../Resources/CSV/crime_data.csv')
harris_df = pd.read_csv('../Resources/CSV/harris_crime_data.csv')
outer_hou_df = pd.read_csv('../Resources/CSV/outer_hou_counties.csv')
```

## Plots

```python
# Apply default sns theme
sns.set_theme(style="darkgrid")

# Crate a visualization
sns.relplot(
    data = crime_df,
    x = "Violent_Offenses", 
    y = "Population",
    hue = "County",
    col = "County",
    col_wrap=3,
)
plt.show(sns)
```
